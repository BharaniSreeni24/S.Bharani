```python
# Step 1: Unzip the dataset file (make sure netflix_titles.csv.zip is uploaded in Colab)
!unzip -o netflix_titles.csv.zip

# Step 2: Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid")

# Step 3: Load the CSV file
df = pd.read_csv('netflix_titles.csv')
print("Dataset shape:", df.shape)
display(df.head())

# Step 4: Explore dataset info and missing values
print("\nData info:")
df.info()
print("\nMissing values per column:")
print(df.isnull().sum())

# Step 5: Data cleaning
df = df.dropna(subset=['country', 'rating', 'duration'])
df['director'].fillna('Unknown', inplace=True)
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
df = df.dropna(subset=['date_added'])  # Remove rows where date_added couldn't convert
df['year_added'] = df['date_added'].dt.year

# Step 6: Plot 1 - Content Type Distribution
plt.figure(figsize=(6,4))
sns.countplot(data=df, x='type', palette='Set2')
plt.title('Content Type Distribution')
plt.xlabel('Type')
plt.ylabel('Count')
plt.show()

# Step 7: Plot 2 - Top 10 Countries by Content Count
plt.figure(figsize=(10,5))
df['country'].value_counts().head(10).plot(kind='bar', color='orange')
plt.title('Top 10 Countries with Most Netflix Content')
plt.xlabel('Country')
plt.ylabel('Number of Titles')
plt.xticks(rotation=45)
plt.show()

# Step 8: Plot 3 - Content Added Over the Years
plt.figure(figsize=(8,4))
df['year_added'].value_counts().sort_index().plot(kind='line', marker='o', color='green')
plt.title('Netflix Content Added Over the Years')
plt.xlabel('Year')
plt.ylabel('Number of Titles Added')
plt.grid(True)
plt.show()

# Step 9: Plot 4 - Top 10 Genres/Categories
plt.figure(figsize=(8,5))
df['listed_in'].value_counts().head(10).plot(kind='barh', color='purple')
plt.title('Top 10 Content Categories')
plt.xlabel('Count')
plt.ylabel('Category')
plt.show()

# Step 10: Plot 5 - Movie Duration Distribution
movies = df[df['type'] == 'Movie'].copy()
# Remove ' min' and convert to int, handle errors if any
movies['duration_minutes'] = movies['duration'].str.replace(' min', '', regex=False).astype(int)

plt.figure(figsize=(8,4))
plt.hist(movies['duration_minutes'], bins=30, color='skyblue', edgecolor='black')
plt.title('Distribution of Movie Durations')
plt.xlabel('Duration (minutes)')
plt.ylabel('Number of Movies')
plt.show()

print("ðŸŽ‰ Project complete! You can now export this notebook or upload it to GitHub.")
```
